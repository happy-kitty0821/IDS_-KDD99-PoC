```python
#!/usr/bin/python3
import pandas
import numpy
```

```python
dataframe = pandas.read_csv("Dataset/kddcup99.csv")
shape = dataframe.shape
numberofrowsfortesting = 115999
#generating random indices for the testigng set
testing_indices = numpy.random.choice(shape[0], numberofrowsfortesting, replace=False)
# Create an array of boolean values to select the training set
training_mask = numpy.ones(shape[0], dtype=bool)
training_mask[testing_indices] = False

# Split the data into training and testing sets
training_data = dataframe[training_mask]
testing_data = dataframe[~training_mask]

# Save training and testing sets to separate CSV files
training_data.to_csv('Dataset/train.csv', index=False)
testing_data.to_csv('Dataset/test.csv', index=False)
```
